```python
import yaml
import pandas as pd
import numpy as np
```

```python
with open("./order.yml") as f:
    draft_order = yaml.load(f, yaml.Loader)
```

```python
draft_order
```

```python
df = pd.read_csv("./data/matches.csv")
```

```python
df
```

```python
def get_draft_ban(row, team):
    team_number = row.index[row == team][0].split('_')[0]
    first_pick = bool(row[team_number+'_pick'])
    side = row[team_number+'_side']
    
    if first_pick:
        order = draft_order['Team1']
    else:
        order = draft_order['Team2']
        
    bans = row[[str(i) for i in order['ban']]]
    picks = row[[str(i) for i in order['pick']]]
    
    return {
        'team': team, 'first_pick': first_pick, 'side': side, 'bans': list(bans), 'picks': list(picks)
    }
```

```python
res_df = []

for i, row in df.iterrows():
    teams = list(row[['team1_name', 'team2_name']])
    for team in teams:
        res = get_draft_ban(row, team)
        res_df.append(res)
```

```python
df = pd.DataFrame(res_df)
```

```python
df
```

```python
df.to_csv("./data/matches_longform.csv", index=False)
```
