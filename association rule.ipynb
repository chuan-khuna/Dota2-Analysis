```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder


from pyvis.network import Network
```

```python
df = pd.read_csv("./data/matches_longform.csv")
```

```python
df['bans'] = df['bans'].apply(lambda l: eval(l))
df['picks'] = df['picks'].apply(lambda l: eval(l))
```

```python
preprocessor = TransactionEncoder()
```

```python
data = preprocessor.fit_transform(list(df['picks']))
df_onehot = pd.DataFrame(data, columns=preprocessor.columns_)
df_onehot = df_onehot.astype(int)
df_onehot
```

```python
data
```

```python
apriori_df = apriori(df_onehot, use_colnames=True, verbose=1, min_support=0.01)
apriori_df
```

```python
df = association_rules(apriori_df, metric="confidence", min_threshold=0.1)
```

```python
df
```

```python
colors = ["#cb0669", "#ce397a", "#cf578a", "#cd6f97", "#c986a2", "#c39aab", "#bcaeb3"][::-1]
node_colors = ["#4757d1", "#2b9be6", "#38d1f0"]
```

```python
def add_node_if_not_exist(net, node, **kwargs):
    if node not in net.get_nodes():
        net.add_node(node, **kwargs)
```

```python
max_items = 1

filtered_df = df[(
                  df['antecedents'].apply(lambda x: len(x) <= max_items) &
                  df['consequents'].apply(lambda x: len(x) <= max_items)
                )]
filtered_df
```

```python
filtered_df['lift_col'] = pd.qcut(filtered_df['lift'], q=len(colors), labels=list(range(len(colors))))
```

```python
node_size_factor = 100
edge_width_factor = 5


net = Network(height='1000px', bgcolor="#222222", font_color="white", select_menu=True, directed=True)
net.force_atlas_2based(spring_length=250)
net.set_edge_smooth('dynamic')
# net.toggle_hide_edges_on_drag(True)
net.show_buttons(filter_=['physics'])
net.inherit_edge_colors(False)

for i, row in filtered_df.iterrows():
    
    ant = list(row['antecedents'])
    con = list(row['consequents'])
    
    if len(ant) - 1 > len(node_colors):
        ant_color = node_colors[-1]
    else:
        ant_color = node_colors[len(ant) - 1]
        
        
    if len(con) - 1 > len(node_colors):
        con_color = node_colors[-1]
    else:
        con_color = node_colors[len(ant) - 1]
    
    ant_title = '\n'.join(ant)
    con_title = '\n'.join(con)
    
    ant_size = row['antecedent support']
    con_size = row['consequent support']
    
    ant_size = np.sqrt(ant_size) * node_size_factor
    con_size = np.sqrt(con_size) * node_size_factor
    
    
    add_node_if_not_exist(net, ant_title, size=ant_size, title=f"{ant_title}\n{np.round(row['antecedent support']*100, 2)}%", color=ant_color)
    add_node_if_not_exist(net, con_title, size=con_size, title=f"{con_title}\n{np.round(row['consequent support']*100, 2)}%", color=con_color)
    
    width = row['confidence'] * edge_width_factor
    color = colors[row['lift_col']]
    
    net.add_edge(ant_title, con_title, width=width, 
                 color=color, 
                 title=f"lift={np.round(row['lift'], 2)}, confidence={np.round(row['confidence'], 3)}")
    
net.show(f"./graphs/arl_{max_items}.html")
```
